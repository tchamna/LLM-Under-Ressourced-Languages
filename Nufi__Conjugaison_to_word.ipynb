```python
import random

import docx
from docx.enum import text as _text, style as _style, table as _table

import json
import requests
import re

import time as _time

# open an existing document
# table_style_list = ["Table Grid",  "Light Table 3 Accent 2", "Light Grid Accent 6", "Light Grid Accent 5", "Light Grid Accent 4", 
#                     "Light Grid Accent 3", "Light Grid Accent 2", "Light Grid Accent 1", "Light Grid"]

table_style_list = ['Light List Accent 2','Light List Accent 4','Light List Accent 5','Light Grid Accent 5']

file_name = 'conjugaison_nufi.docx'

path_verbs = "/content/drive/MyDrive/Data_Science/DataSets/data_verb.json"


nb_table_per_page = 4

# We set the requirements
requirements = {
    # category
    'simple': {
        # group
        1: {
            # x_syllabes: number of syllabes
            1: {
                # constraints: need
                'mb': 5, 'nd': 5, 'nj': 5, 'ng': 5, 'ngw': 5,
            },
            2: {
                'mb': 5, 'nd': 5, 'nj': 5, 'ng': 5, 'ngw': 5,
            },
            3: {
                'mb': 5, 'nd': 5, 'nj': 5, 'ng': 5, 'ngw': 5,
            }
        },
        2: {
            1: {
                'mb': 5, 'nd': 5, 'nj': 5, 'ng': 5, 'ngw': 5,
            },
            2: {
                'mb': 5, 'nd': 5, 'nj': 5, 'ng': 5, 'ngw': 5,
            },
            3: {
                'mb': 5, 'nd': 5, 'nj': 5, 'ng': 5, 'ngw': 5,
            }
        },
    },
    'compose': {
        1: {
            1: {
                'mb': 5, 'nd': 5, 'nj': 5, 'ng': 5, 'ngw': 5,
            },
            2: {
                'mb': 5, 'nd': 5, 'nj': 5, 'ng': 5, 'ngw': 5,
            },
            3: {
                'mb': 5, 'nd': 5, 'nj': 5, 'ng': 5, 'ngw': 5,
            }
        },
        2: {
            1: {
                'mb': 5, 'nd': 5, 'nj': 5, 'ng': 5, 'ngw': 5,
            },
            2: {
                'mb': 5, 'nd': 5, 'nj': 5, 'ng': 5, 'ngw': 5,
            },
            3: {
                'mb': 5, 'nd': 5, 'nj': 5, 'ng': 5, 'ngw': 5,
            }
        },
    },
}

# We set the time groups
# The index should respect the order of the nufi conugator
time_groups = {
    'Présent': {
        0: 'Présent',
        10: 'Impératif'
    },
    'Passé Récent': {
        1: 'Passé Composé',
        2: 'Passé Récent',
    },
    'Passé': {
        3: 'Passé d\'hier',
        4: 'Passé (>= 2 jours)',
        6: 'Passé Lointain',
    },
    'Futur Proche': {
        7: 'Futur Proche',
    },
    'Futur Lointain': {
        8: 'Futur Lointain',
    },
    'Présent Habituel': {
        9: 'Présent Habituel',
    },
}

def insert_table(document: docx.document.Document, headers: list, data: list, title: str) -> None:
    heading = document.add_heading(title, level=1)
    heading.alignment = _text.WD_ALIGN_PARAGRAPH.CENTER
    heading.bold = False

    rows = len(data)
    columns = len(headers)

    # add a table to the end and create a reference variable
    # extra row is so we can add the header row
    table = document.add_table(rows+1, columns)
    table.alignment = _table.WD_TABLE_ALIGNMENT.CENTER

    # Adding style to a table
    table.style = random.choice(table_style_list)
    table.allow_autofit = True
    table.autofit = True

    # add the header rows
    for j in range(columns):
        cell = table.cell(0,j)
        pararagraph = cell.paragraphs[0]
        pararagraph.alignment = _text.WD_ALIGN_PARAGRAPH.CENTER
        pararagraph.add_run(headers[j])
        pararagraph.style='Nufi'
        pararagraph.bold = False

    # add the rest of the data
    for i in range(rows):
        for j in range(columns):
            cell = table.cell(i+1, j)
            pararagraph = cell.paragraphs[0]
            pararagraph.style = 'Nufi'
            pararagraph.alignment = _text.WD_ALIGN_PARAGRAPH.CENTER
            pararagraph.add_run(data[i][j]).bold = False

def get_conjugaison(verb: str) -> list:
    # html = open('/tmp/untitled.html', 'r').read()
    html = requests.get(f"https://www.resulamdev.com/nufi-conjugator/?verb={verb}&valider=#comment-391").text
    # html = requests.get(f"https://resulam.com/nufi-conjugator?verb={verb}").text
    html = html.replace('\n','')
    
    # # We get the conjugaison by time
    # conj = re.findall(r'<tbody>(.*?)</tbody>', html)
    # # We get the conj of the same time
    # conj = [re.findall(r'<tr.*?>(.*?)</tr>', i) for i in conj]
    # # We split the conj of the same time
    # conj = [[re.findall(r'<td.*?>(.*?)</td>', ii) for ii in i] for i in conj]
    # # We arrange the data
    # conj = [[[iii.strip() for iii in ii] for ii in i] for i in conj]

    # We get the conjugaison
    conj = [
        [
            [
                # We arrange the data
                iii.strip()
                # We split the conj of the same time
                for iii in re.findall(r'<td.*?>(.*?)</td>', ii)
            ]
            # We get the conj of the same time
            for ii in re.findall(r'<tr.*?>(.*?)</tr>', i)
        ]
        # We get the conjugaison by time
        for i in re.findall(r'<tbody>(.*?)</tbody>', html)
    ]

    return conj

print('Loading data...')
with open(path_verbs, 'r') as f:
    data_verb = json.load(f)

print('Configuration document...')
# We config the document
document = docx.Document()
style = document.styles.add_style('Nufi', _style.WD_STYLE_TYPE.PARAGRAPH)
style.font.size = docx.shared.Pt(12)
style.font.name = 'Charis-SIL'
style.font.bold = False

print('Building documents...')
for category, groups in requirements.items():
    document.add_page_break()
    document.add_heading(category, level=0)

    for group, x_syllabes in groups.items():
        document.add_page_break()
        document.add_heading(f'Group {group}', level=0)

        for nb_syllabe, needs in x_syllabes.items():
            document.add_page_break()
            document.add_heading(f'{nb_syllabe} syllabe(s)', level=0)

            for constraint in needs:
                document.add_page_break()
                document.add_heading(f'Verb start with {constraint}', level=0)

                print(f'\tcategory: {category}')
                print(f'\tgroup: {group}')
                print(f'\tnb_syllabe: {nb_syllabe}')
                print(f'\tconstraint: {constraint}')

                for d in data_verb:
                    # We stop if the need is sastify
                    if needs[constraint] <= 0:
                        break

                    nb_table = 0
                    verb = d['verb']
                    translated_verb = d['translated_verb']

                    if not verb.startswith(constraint):
                        continue

                    # We remove the verb from the list
                    data_verb.remove(d)
                    
                    print(f'\t\tGetting conjugaison of {verb}...')

                    delay = 1
                    while True:
                        try:
                            _time.sleep(delay) # A delay to evict to overload the server
                            conj = get_conjugaison(verb)
                            break
                        except:
                            delay += 1
                            print(f'\t\tRetry with a delay of {delay}s...')
                            _time.sleep(delay)

                    # We verify if the conjugaison is found
                    if not conj:
                        continue

                    # We update the need
                    needs[constraint] -= 1;

                    print(f'\t\tBuilding conjugaison of {verb}...')

                    for title, time_group in time_groups.items():

                        headers = []
                        data = []

                        for time, subtitle in time_group.items():
                            # We build the header in function of the number of sub conj
                            if len(conj[time][0]) > 1:
                                headers += [f'{subtitle} {i+1}' for i in range(len(conj[time][0]))]
                            else:
                                headers.append(subtitle)

                            # We merge the conj of time of this group
                            if data == []:
                                data = conj[time]
                            else:
                                for i in range(len(conj[time])):
                                    data[i] += conj[time][i]
                                # endfor

                        if nb_table % nb_table_per_page == 0:
                            document.add_page_break()
                        else:
                            pass

                        insert_table(document, headers, data, f"{title} ({verb}: {translated_verb})")
                        nb_table += 1
                    # endfor
                # endfor
            # endfor
            print('\tSaving...')
            document.save(file_name)
        # endfor
    # endfor
# endfor

print("Finish")
```
